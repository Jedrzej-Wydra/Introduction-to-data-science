```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# f(x) = x**2 - 2
# f'(x) = 2x

def f(x):
  return x**2 - 2

def df(x):
  return 2*x

x0 = 2
y0 = f(x0)

# f(x) = x**2 - 2
# f'(x) = 2*x
# y - y0 = f'(x0)(x - x0)
# 0 - y0 = f'(x0)(x - x0)
# f'(x0)x0 - y0 = f'(x0)(x)
# x = (f'(x0)x0 - y0)/f'(x0)
# x = x0 - y0/f'(x0)

x = []
iter = 5

for i in range(iter):
  x.append(x0 - y0/df(x0))
  x0 = x[-1]
  y0 = f(x0)

plt.plot(list(range(iter)), x)
plt.plot(list(range(iter)), [np.sqrt(2)]*iter)
plt.show()

plt.plot(list(range(iter)), [abs(i - np.sqrt(2)) for i in x])
plt.show()
```

```python
# f(x) = e**x - 10
# f'(x) = e**x

def factorial(x):
  res = 1
  for i in range(1, x + 1):
    res *= i
  return res

def f(x, n):
  result = 0
  for i in range(n):
    result += (x**i)/(factorial(i))

  return result - 10

def df(x, n):
  return f(x, n) + 10


approx = 100
x0 = 2
y0 = f(x0, approx)

# y - y0 = f'(x0)(x - x0)
# 0 - y0 = f'(x0)(x - x0)
# f'(x0)x0 - y0 = f'(x0)(x)
# x = (f'(x0)x0 - y0)/f'(x0)
# x = x0 - y0/f'(x0)
# e**x = sum_k((x**k)/k!)

x = []
iter = 5

for i in range(iter):
  x.append(x0 - y0/df(x0, approx))
  x0 = x[-1]
  y0 = f(x0, approx)

plt.plot(list(range(iter)), x)
plt.plot(list(range(iter)), [np.log(10)]*iter)
plt.show()

plt.plot(list(range(iter)), [abs(i - np.log(10)) for i in x])
plt.show()
```

```python
def pi(n_rand):
  x = [np.random.uniform(-1, 1) for i in range(n_rand)]
  y = [np.random.uniform(-1, 1) for i in range(n_rand)]

  inside = [xi**2 + yi**2 < 1 for xi, yi in zip(x, y)]

  result = 4*sum(inside)/len(inside)

  return result

def pi_mean(n_rand, n):
  res = [pi(n_rand) for i in range(n)]
  return sum(res)/n

n_rand = 100

x = [np.random.uniform(-1, 1) for i in range(n_rand)]
y = [np.random.uniform(-1, 1) for i in range(n_rand)]

plt.figure(figsize=(6, 6))
plt.plot([-1,1], [1,1], color = 'steelblue')
plt.plot([-1,1], [-1,-1], color = 'steelblue')
plt.plot([-1,-1], [-1,1], color = 'steelblue')
plt.plot([1,1], [-1,1], color = 'steelblue')
plt.scatter(x, y, color = 'coral')
xx = np.linspace(-1,1,100)
plt.plot(xx, np.sqrt(1 - xx**2), color = 'steelblue')
plt.plot(xx, -np.sqrt(1 - xx**2), color = 'steelblue')
plt.show()
print(pi(n_rand))
```

```python
val_pi = []

iter = 240
n_rand = 100
for i in range(1, iter + 1):
  val_pi.append(pi_mean(n_rand = i, n = 100))

plt.plot(list(range(iter)), val_pi)
plt.plot(list(range(iter)), [np.pi]*(iter))
plt.show()
```

Z danych z zadania drugiego proszę utworzyć klasę, która przy tworzeniu obiektu jako argument przyjmuje słownik z danymi, a utworzony obiekt zawiera metodę:
1. zwracającą nazwiska w kolejności alfabetycznej;
2. ponumerowaną listę zawodów;
3. nieuporządkowane imiona;
4. zwracającą informacje o danych przechowywanych przez obiekt.

Obiekt ma zawierać przynajmniej trzy metody.

```python
class profesions_class:
  def __init__(self, dict_arg):
    self.ans = dict_arg

  def sorted_names(self):
    self.result = sorted([i[0].split(' ')[1] for i in self.ans.items()])
    return self.result

  def numbered_profesions(self):
    self.result = []
    self.helper_list = list(self.ans.values())
    for i in range(len(self.helper_list)):
      self.result.append((i + 1, self.helper_list[i]))
    return self.result

  def unordered_names(self):
    self.helper_list = [i[0].split(' ')[0] for i in self.ans.items()]
    self.result = set(self.helper_list)
    return self.result

  def __str__(self):
    lines = []
    for name, job in self.ans.items():
      lines.append(f"imię i nazwisko: {name}\nzawód: {job}\n")
    return "\n".join(lines)
```

## Zadanie 13
Proszę zdefiniować pythonową funkcję, która rozwiązuje równianie kwadratowe postaci $ax^2 + bx + c = 0$ bez używania Numpy.

```python
def solve_quadratic(a, b, c):
    if a == 0 and b == 0 and c == 0:
        return "Nieskończenie wiele rozwiązań"
    if a == 0 and b != 0:
        return (-c/b, )
    if a == 0 and b == 0 and c != 0:
        return "Brak rozwiązań"
    delta = b**2 - 4*a*c
    if delta < 0:
        return "Brak rozwiązań rzeczywistych"
    sqrt_delta = delta**0.5
    if delta == 0:
        x0 = -b/(2*a)
        return (x0, )
    else:
        x1 = (-sqrt_delta-b)/(2*a)
        x2 = (sqrt_delta-b)/(2*a)
        if x1 < x2:
            return (x1, x2)
        else:
            return (x2, x1)

solve_quadratic(1, 1, -1)
```

## Zadanie 18
Proszę zdefiniować pythonową funkcję, która sprawdza, czy podana liczba jest liczbą pierwszą.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

is_prime(19)
```

```python
is_prime(4)
```
